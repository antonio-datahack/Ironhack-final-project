```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
from tqdm.notebook import tqdm
import numpy as np

from string import ascii_lowercase
from random import uniform
from time import sleep
```

```python
url = "https://www.operabase.com/managers/en"
response = requests.get(url)
html= response.content
scraped = soup(html, "html.parser")
```

```python
scraped
```

```python
countr = scraped.select("#__next > div.jss15 > div.jss20 > div > div > div.jss10 > ul > li > a")
len(countr)
```

```python
name_country = []
for i in range(len(countr)):
    name_country.append(countr[i]["href"])
name_country
```

```python
countries = []
for country in name_country:
    countries.append(country[10:-3])
countries
```

```python
from random import uniform
pages = []
url_lst = []
counter = 0

for country in countries: 
        url = "https://www.operabase.com/managers/"+ country.strip("-").lower() +"/en"
        print(url)
        url_lst.append(url)
        counter +=1
        # start_at needs to be a string if we used this:
        # url = "https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1992-12-31&user_rating=7.5,&start=" + start_at + "&ref_=adv_nxt"

        # download the url
        response = requests.get(url)

        # print the status code
        print("Status code: " + str(response.status_code))
        print(counter)
        pages.append(response)

        # respectful nap
        wait_time = uniform(0.3, 2)
        sleep(wait_time)
    
url_lst
```

```python
pages_countries = pd.DataFrame(pages) 
    
# saving the dataframe 
pages_countries.to_csv('../data/pages_countries.csv')
```
